```python
import pandas as pd
import os
import glob
```

```python
# Concatenate monthly tenders files
as_tenders_root = r"D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\Deployment\flood-data-ecosystem-Assam\Sources\TENDERS\data\monthly_tenders"
if not os.path.isdir(as_tenders_root):
    raise FileNotFoundError(f"Directory not found: {as_tenders_root}")

csv_files = glob.glob(os.path.join(as_tenders_root, "*.csv"))
frames = []
for file in csv_files:
    try:
        df = pd.read_csv(file)
        frames.append(df)
    except Exception as e:
        print(f"Failed to read {file}: {e}")

if frames:
    as_tenders = pd.concat(frames, ignore_index=True)
else:
    as_tenders = pd.DataFrame()

as_tenders.shape
```
