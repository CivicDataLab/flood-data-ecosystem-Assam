```python
#root="/Users/saurabhlevin/Deployment"

root = "D:\CivicDataLab_IDS-DRR\IDS-DRR_Github/Deployment/flood-data-ecosystem-Assam/"
SEC_51 = pd.read_csv(root+r'Sources/TENDERS/data/SDRF/SEC/extracted/51st_SEC_flood_related_allocations_aggregated.csv')
SEC_52 = pd.read_csv(root+r'Sources/TENDERS/data/SDRF/SEC/extracted/52nd_SEC_flood_related_allocations_aggregated.csv')
SEC_53 = pd.read_csv(root+r'Sources/TENDERS/data/SDRF/SEC/extracted/53rd_SEC_flood_related_allocations_aggregated.csv')
SEC_54 = pd.read_csv(root+r'Sources/TENDERS/data/SDRF/SEC/extracted/54th_SEC___Aggregated_totals_by_district__from_extracted_items_.csv')
```

```python
concat = pd.concat([SEC_51, SEC_52, SEC_53, SEC_54], ignore_index=True)
# convert 'date' to YYYY_MM (e.g. 2024_10)
concat['date'] = pd.to_datetime(concat['date'], errors='coerce').dt.strftime('%Y_%m')
concat['total_amount_lakhs'] = concat['total_amount_lakhs']*100000
```

```python
concat['district'] = concat['district'].replace({
    'Kamrup': 'Kamrup Metro',
    'Kamrup (rural)': 'Kamrup',
    'Kamrup (Rural)': 'Kamrup'
})
concat['district'] = concat['district'].str.upper()
concat = concat.rename(columns={'district':'District','date':'timeperiod','total_amount_lakhs':'SDRF funding'})
```

```python
concat.to_csv(root+r'/Sources/TENDERS/data/SDRF/SEC/extracted/51_to_54th_SEC_flood_related_allocations_aggregated.csv', index=False)
```

### Correlation analysis

```python
from scipy.stats import pearsonr
```

```python

advance_sdrf = pd.read_csv(r'/Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/SDRF/SEC/extracted/54th_SEC___Advance_SDRF_allocations__Annexure_VIII__ranked___CORRECTED.csv')
advance_sdrf['district'] = advance_sdrf['district'].str.upper()
```

```python
advance_sdrf['district'] = advance_sdrf['district'].replace({'KARBI': 'KARBI ANGLONG', 'DIMAHASAO': 'DIMA HASAO','SOUTH':'SOUTH SALMARA MANCACHAR'})
```

```python
gd_2024 = pd.read_excel(r'/Users/saurabhlevin/Downloads/Report on IDS-DRR _ SDRF Insights and Pre-Monsoon Training_supporting-data.xlsx', sheet_name='2024_GD-Sir')
method2_2024 = pd.read_excel(r'/Users/saurabhlevin/Downloads/Report on IDS-DRR _ SDRF Insights and Pre-Monsoon Training_supporting-data.xlsx', sheet_name='2024_method2')
ids_2024 = pd.read_excel(r'/Users/saurabhlevin/Downloads/Report on IDS-DRR _ SDRF Insights and Pre-Monsoon Training_supporting-data.xlsx', sheet_name='2024_ids-drr')

gd_hist = pd.read_excel(r'/Users/saurabhlevin/Downloads/Report on IDS-DRR _ SDRF Insights and Pre-Monsoon Training_supporting-data.xlsx', sheet_name='Historical_GD-Sir')
method2_hist = pd.read_excel(r'/Users/saurabhlevin/Downloads/Report on IDS-DRR _ SDRF Insights and Pre-Monsoon Training_supporting-data.xlsx', sheet_name='Historical_method2')
ids_hist = pd.read_excel(r'/Users/saurabhlevin/Downloads/Report on IDS-DRR _ SDRF Insights and Pre-Monsoon Training_supporting-data.xlsx', sheet_name='Historical_ids-drr')
```

```python
merged_ids2024 = ids_2024.merge(advance_sdrf, on='district', how='outer', suffixes=('_ids', '_method2'))
p1, r1= pearsonr(merged_ids2024['advance_sdrf_allocation'].dropna(), merged_ids2024['Allocation factor'][merged_ids2024['advance_sdrf_allocation'].notna()])
print(f"ids2024 Pearson correlation coefficient: {p1}")
print(f"P-value: {r1}")
```

```python
merged_idshist = ids_hist.merge(advance_sdrf, on='district', how='outer', suffixes=('_ids', '_method2'))
p1, r1= pearsonr(merged_idshist['advance_sdrf_allocation'].dropna(), merged_idshist['Allocation Factor'][merged_idshist['advance_sdrf_allocation'].notna()])
print(f"idshist Pearson correlation coefficient: {p1}")
print(f"P-value: {r1}")
```

```python
merged_method2 = method2_2024.merge(advance_sdrf, left_on='District', right_on='district', how='outer', suffixes=('_ids', '_method2'))
merged_method2 = merged_method2.dropna(subset=['advance_sdrf_allocation', 'Allocation Factor'])
p2, r2 = pearsonr(merged_method2['advance_sdrf_allocation'].dropna(), merged_method2['Allocation Factor'][merged_method2['advance_sdrf_allocation'].notna()])
print(f"Pearson correlation coefficient: {p2}")
print(f"P-value: {r2}")
```

```python
merged_method2_hist = method2_hist.merge(advance_sdrf, left_on='district', right_on='district', how='outer', suffixes=('_ids', '_method2'))
merged_method2_hist = merged_method2_hist.dropna(subset=['advance_sdrf_allocation', 'allocation_share'])
p2, r2 = pearsonr(merged_method2_hist['advance_sdrf_allocation'].dropna(), merged_method2_hist['allocation_share'][merged_method2_hist['advance_sdrf_allocation'].notna()])
print(f"Pearson correlation coefficient: {p2}")
print(f"P-value: {r2}")
```

```python
merged_gdmethod = gd_2024.merge(advance_sdrf, left_on='District', right_on='district', how='outer', suffixes=('_ids', '_method2'))
p3, r3 = pearsonr(merged_gdmethod['advance_sdrf_allocation'].dropna(), merged_gdmethod['Allocation Factor'][merged_gdmethod['advance_sdrf_allocation'].notna()])
print(f"Pearson correlation coefficient: {p3}")
print(f"P-value: {r3}")
```

```python
merged_gd_hist = gd_hist.merge(advance_sdrf, left_on='district', right_on='district', how='outer', suffixes=('_ids', '_method2'))
p3, r3 = pearsonr(merged_gd_hist['advance_sdrf_allocation'].dropna(), merged_gd_hist['allocation_share'][merged_gd_hist['advance_sdrf_allocation'].notna()])
print(f"Pearson correlation coefficient: {p3}")
print(f"P-value: {r3}")
```
