```python
import pandas as pd
import numpy as np
```

```python
SEC_51 = pd.read_csv(r'/Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/SDRF/SEC/extracted/51st_SEC_flood_related_allocations_aggregated.csv')
SEC_52 = pd.read_csv(r'/Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/SDRF/SEC/extracted/52nd_SEC_flood_related_allocations_aggregated.csv')
SEC_53 = pd.read_csv(r'/Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/SDRF/SEC/extracted/53rd_SEC_flood_related_allocations_aggregated.csv')
SEC_54 = pd.read_csv(r'/Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/SDRF/SEC/extracted/54th_SEC___Aggregated_totals_by_district__from_extracted_items_.csv')
```

```python
SEC_54
```

```python
concat = pd.concat([SEC_51, SEC_52, SEC_53], ignore_index=True)
# convert 'date' to YYYY_MM (e.g. 2024_10)
concat['date'] = pd.to_datetime(concat['date'], errors='coerce').dt.strftime('%Y_%m')
concat['total_amount_lakhs'] = concat['total_amount_lakhs']*100000
```

```python
concat['district'] = concat['district'].replace({
    'Kamrup': 'Kamrup Metro',
    'Kamrup (rural)': 'Kamrup',
    'Kamrup (Rural)': 'Kamrup'
})
concat['district'] = concat['district'].str.upper()
concat = concat.rename(columns={'district':'District','date':'timeperiod','total_amount_lakhs':'SDRF funding'})
```

```python
concat.to_csv(r'/Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/SDRF/SEC/extracted/51_to_53rd_SEC_flood_related_allocations_aggregated.csv', index=False)
```
